```python
import numpy as np
import matplotlib.pyplot as plt
```

# Равномерное распределение

```python
from scipy.stats import uniform

a1 = 0
n1 = 5

a2 = 2
n2 = 4
```

```python
x = np.linspace(-3, 8, 1000)
y1 = uniform.pdf(x, a1, n1)
y2 = uniform.pdf(x, a2, n2)

plt.plot(x, y1, color='blue', label=f'U(a={a1}, b={a1+n1})')
plt.plot(x, y2, color='red', label=f'U(a={a2}, b={a2+n2})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Плотность равномерного распределения')
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.linspace(-3, 8, 1000)
y1 = uniform.cdf(x, a1, n1)
y2 = uniform.cdf(x, a2, n2)

plt.plot(x, y1, color='blue', label=f'U(a={a1}, b={a1+n1})')
plt.plot(x, y2, color='red', label=f'U(a={a2}, b={a2+n2})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция равномерного распределения')
plt.legend()
plt.grid()
plt.show()
```

# Биномиальное распределение

```python
from scipy.stats import binom

n1 = 25
p1 = 0.6

n2 = 10
p2 = 0.5
```

```python
x = np.linspace(0, 29, 30)
y1 = binom.pmf(x, n1, p1)
y2 = binom.pmf(x, n2, p2)

plt.scatter(x, y1, color='blue', label=f'B(n={n1}, p={p1})')
plt.scatter(x, y2, color='red', label=f'B(n={n2}, p={p2})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция вероятности биномиального распределения')
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.linspace(0, 29, 30)
y1 = binom.cdf(x, n1, p1)
y2 = binom.cdf(x, n2, p2)

plt.scatter(x, y1, color='blue', label=f'B(n={n1}, p={p1})')
plt.scatter(x, y2, color='red', label=f'B(n={n2}, p={p2})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция биномиального распределения')
plt.legend()
plt.grid()
plt.show()
```

# Распределение Пуассона

```python
from scipy.stats import poisson

mu1 = 1.5

mu2 = 0.5
```

```python
x = np.linspace(-2, 6, 9)
y1 = poisson.pmf(x, mu1)
y2 = poisson.pmf(x, mu2)

plt.scatter(x, y1, color='blue', label=f'P(mu={mu1})')
plt.scatter(x, y2, color='red', label=f'P(mu={mu2})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция вероятности распределения Пуассона')
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.linspace(-2, 6, 9)
y1 = poisson.cdf(x, mu1)
y2 = poisson.cdf(x, mu2)

plt.scatter(x, y1, color='blue', label=f'P(mu={mu1})')
plt.scatter(x, y2, color='red', label=f'P(mu={mu2})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция распределения Пуассона')
plt.legend()
plt.grid()
plt.show()
```

# Нормальное распределение

```python
from scipy.stats import norm

mu1 = 1.0
sigma1 = 2.0

mu2 = 3.0
sigma2 = 1.0
```

```python
x = np.linspace(-5, 7, 1000)
y1 = norm.pdf(x, mu1, sigma1)
y2 = norm.pdf(x, mu2, sigma2)

plt.plot(x, y1, color='blue', label=f'N(mu={mu1}, sigma={sigma1})')
plt.plot(x, y2, color='red', label=f'N(mu={mu2}, sigma={sigma2})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Плотность нормального распределения')
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.linspace(-5, 7, 1000)
y1 = norm.cdf(x, mu1, sigma1)
y2 = norm.cdf(x, mu2, sigma2)

plt.plot(x, y1, color='blue', label=f'N(mu={mu1}, sigma={sigma1})')
plt.plot(x, y2, color='red', label=f'N(mu={mu2}, sigma={sigma2})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция нормального распределения')
plt.legend()
plt.grid()
plt.show()
```
